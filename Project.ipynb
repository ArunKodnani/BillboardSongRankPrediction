```python
from bs4 import BeautifulSoup
import requests
import pandas as pd

def scrapBillboard():
    data_billboard = pd.DataFrame(columns=['Rank','Song Name','Artist'])
    page = requests.get("http://www.billboard.com/charts/billboard-200")
    soup = BeautifulSoup(page.content, 'html.parser')
    for i in range(1,201):
        song = soup.find('article', class_= 'chart-row chart-row--'+str(i)+' js-chart-row')
        songRank = song.find(class_= "chart-row__current-week").get_text()
        songName = song.find(class_= "chart-row__song").get_text()
        songArtist = song.find(class_= "chart-row__artist").text.strip('\n')
        data_billboard.loc[i] = [songRank, songName, songArtist]
    return data_billboard.set_index('Rank')

scrapBillboard()
#[type(item) for item in list(html.children)] 
```
